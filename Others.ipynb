```python
import numpy as np
from sklearn.cluster import KMeans
import cv2
from matplotlib import pyplot as plt


def kmeanscolors(n_colors, image='balloons.jpg'):
    im = cv2.imread(image)
    h,w = im.shape[:2]
    im = im.reshape(h*w, 3)
    kmeans = KMeans(n_clusters=n_colors, init='k-means++')
    kmeans.fit(im)
    for i, pred in enumerate(kmeans.predict(im)):
        im[i] = kmeans.cluster_centers_[pred]
    newim = im.reshape(h, w, 3)
    plt.imshow(cv2.cvtColor(newim, cv2.COLOR_BGR2RGB))
```

```python
kmeanscolors(5, image='balloons.jpg')
```
